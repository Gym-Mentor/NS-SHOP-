```python
df_month = df[["month", "uniquecode", "quantity"]]
```

```python
df_month.columns = ["month", "uniquecode", "quantity"]
```

```python
df_spring = df_month[(df_month["month"] == 3) | (df_month["month"] == 4) | (df_month["month"] == 5)]
df_spring = pd.DataFrame(df_spring[["uniquecode","quantity"]].groupby(["uniquecode"]).agg("mean").reset_index())
```

```python
df_summer = df_month[(df_month["month"] == 6) | (df_month["month"] == 7) | (df_month["month"] == 8)]
df_summer = pd.DataFrame(df_summer[["uniquecode","quantity"]].groupby(["uniquecode"]).agg("mean").reset_index())
```

```python
df_fall = df_month[(df_month["month"] == 9) | (df_month["month"] == 10) | (df_month["month"] == 11)]
df_fall = pd.DataFrame(df_fall[["uniquecode","quantity"]].groupby(["uniquecode"]).agg("mean").reset_index())
```

```python
df_winter = df_month[(df_month["month"] == 12) | (df_month["month"] == 1) | (df_month["month"] == 2)]
df_winter = pd.DataFrame(df_winter[["uniquecode","quantity"]].groupby(["uniquecode"]).agg("mean").reset_index())
```
