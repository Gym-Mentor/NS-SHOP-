```python
old =""
rating = []
total_rating = 0
for i in range(slicing_df.shape[0]):
    sc =slicing_df.iloc[i]
    if sc["playtime"] !=0:
        if i ==0:
            total_rating = total_rating +sc["rating"]
        else:
            rating.append(total_rating)
            total_rating = 0
            total_rating = total_rating +sc["rating"]
    else:
        total_rating = total_rating +sc["rating"]

    # 맨끝 열은 12-31 23:59분으로 끝나서 이 해당 데이터까지 저장 후 종료.    
    if i == 459719:
        rating.append(total_rating)
```
