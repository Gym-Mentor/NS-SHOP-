```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os,glob,tqdm
from ipywidgets import interact
import warnings

plt.style.use('ggplot')
warnings.filterwarnings('ignore')
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
df=pd.read_excel("./Data/weather_data.xlsx")
print(df.shape)
df.head()
```

```python
df_a=df[['일시','기온(°C)','습도(%)']]
```

```python
#THI=(9/5)*T-0.55*(1-RH)*((9/5)*T-26)+32
arr=[]
for i in df_a.index:
    T=df_a.loc[i,'기온(°C)'] #기온
    RH=df_a.loc[i,'습도(%)']*0.01 #습도 *0.01 
    THI=(9/5)*T-0.55*(1-RH)*((9/5)*T-26)+32 #불쾌지수 공식
    THI = float("{0:.2f}".format(THI)) #소수두째자리까지만 출력
    arr.append(THI)
```

```python
df_a['불쾌지수']=arr
```

```python
for i in df_a.index:
    mon=df_a.loc[i,'일시'].month
    year=df_a.loc[i,'일시'].year
    day=df_a.loc[i,'일시'].day
    hour=df_a.loc[i,'일시'].hour
    df_a.loc[i,'year']=year
    df_a.loc[i,'month']=mon
    df_a.loc[i,'day']=day
    df_a.loc[i,'hour']=hour
```

```python
#2020년도 데이터 삭제
np.where(df_a['year']==2020)
```

```python
df_a=df_a.drop(8755)
```
